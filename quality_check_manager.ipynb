```python
import pandas as pd
```

```python
config_dict = {
    "sap": {
        "quality object": [
            {
                "function_name": "quality_func_in",
                "kw_params": {"column_name": "quality_object", "values": ["UE_RECL", "UE_CONDR", "UE_POLE", "UE_CTOUT", "UE_SWTCH", "UE_VAULP", "UE_CTOTS", "UE_DISC", "UE_XFMER", "UE_RECLS", "UE_RATIO", "UE_REG", "UE_CAPSG" ]}
            }
        ]
    }
}

leer_sap = config_dict.get("sap")
```

```python
def load_dataframe():
  df = pd.DataFrame.from_dict({"quality_object": ["UE_CONDR", "UE_CONDRE", "UE_POLE", "UE_GDF"]})
  return df
```

```python
def get_function_obj(func_name_):
    return globals().get(func_name_, None)
```

```python
def quality_func_in(df_, column_name, values):
    df_valid = df_[df_[column_name].isin(values)]
    df_invalid = df_[~df_[column_name].isin(values)]
    return df_valid, df_invalid
```

```python
def execute_function(df_, function_name, kw_params):
    function_obj = get_function_obj(function_name)
    if function_obj is None:
        Exception("Function " + function_name + " not found.")

    df_v, df_iv = function_obj(df_, **kw_params)
    return df_v, df_iv
```

```python
df = load_dataframe()
df_v = df.copy()
inv_dfs = []
for column_identifier, functions_to_apply in leer_sap.items():
    for function_configuration in functions_to_apply:
        df_v, df_iv = execute_function(df_v, **function_configuration)
        inv_dfs.append(df_iv)
df_inv = pd.concat(inv_dfs)
```
